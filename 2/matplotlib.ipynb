```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,3])
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.show()
```

```python
plt.plot([1,3,5,8],[2,4,5,4])
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.grid()
plt.show()
```

```python
plt.plot([1,3,5,8],[2,4,5,4], 'ro')
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.grid()
plt.show()
```

```python
vals = np.arange(1,5,0.2)
plt.plot(vals, vals**2, "r--", label="^2")
plt.plot(vals, vals**2.2, "bs", label="^2.2")
plt.plot(vals, vals**2.4, "g^", label="^2.4")
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot([1,3,5,8],[2,4,5,4], linewidth=5.0)
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.grid()
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [2,4,6,8]
x2 = x1
y2 = [1,4,9,16]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0], color='r', linewidth=0.5)
plt.setp(lines[1], color='g', linewidth=1.5)
plt.xlabel("Indices")
plt.ylabel("Numbers")
plt.grid()
plt.show()
```

```python
def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0,5.0,0.1)
t2 = np.arange(0,5.0,0.02)

plt.grid()
plt.figure(1)
plt.subplot(211)
plt.plot(t1,f(t1),'b--')

plt.subplot(212)
plt.plot(t2,f(t2),'r-')
plt.show()
```
