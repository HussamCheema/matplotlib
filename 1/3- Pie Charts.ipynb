```python
from matplotlib import pyplot as plt
```

```python
x = plt.style.available
print(x)
```

```python
plt.style.use("fivethirtyeight") # Applying default styles
```

```python
slices = [120,80, 30, 20]
labels = ["Sixty", "Forty", "Extra1", "Extra2"]
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']
```

```python
# wedgeprops, edgecolor is used is border color
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'black'})
plt.title("My awesome pie chart")
plt.tight_layout()
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
```

##### PIE Charts are good when you compare only 5 items
##### They're not good when you compare a lot of items

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})
plt.title("My awesome pie chart")
plt.tight_layout()
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0] # Giving more emphasis to "Python" item
```

```python
plt.pie(slices, labels=labels, explode=explode, shadow=True, 
        startangle=90, wedgeprops={'edgecolor':'black'})
plt.title("My awesome pie chart")
plt.tight_layout()
plt.show()
```

##### To show the percentages

```python
plt.pie(slices, labels=labels, explode=explode, shadow=True, 
        startangle=90, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})
plt.title("My awesome pie chart")
plt.tight_layout()
plt.show()
```
